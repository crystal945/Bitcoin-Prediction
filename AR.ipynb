```python
import time

def convert_time(tt):
    ans=[]
    for t in tt:
        t=t/1000+0.000
        st = time.localtime(t)
        ans.append(time.strftime('%Y-%m-%d %H:%M:%S', st))
    return ans
#convert_time(1513280940000)
```

```python
date = oneday['MTS'].values
date_n = convert_time(date)
```
